```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def sinplot():
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * 0.5) * (7 - i))
```

```python
sinplot()
```

```python
# sns.set() resets everything
sns.set()
sinplot()
```

```python
sns.set_style("whitegrid")
sinplot()
```

```python
sns.set_style("dark")
sinplot()
```

```python
sns.set_style("white")
sinplot()
```

```python
sns.set_style("ticks")
sinplot()
```
