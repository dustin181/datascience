# Plotting with Categorical Data

```python
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set(style="ticks", color_codes=True)
```

## Categorical scatterplots

```python
tips = sns.load_dataset("tips")
tips
```

```python
sns.catplot(x="day", y="total_bill",
            data=tips)
plt.show()
```

```python
sns.catplot(x="day", y="total_bill", 
            jitter=False, data=tips)
plt.show()
```

```python
sns.catplot(x="day", y="total_bill", 
            jitter=True, data=tips)
plt.show()
```

```python
sns.catplot(x="day", y="total_bill", 
            kind="swarm", data=tips)
plt.show()
```

```python
sns.catplot(x="day", y="total_bill", hue="sex", 
            kind="swarm", data=tips)
plt.show()
```

```python
sns.catplot(x="size", y="total_bill", kind="swarm",
            data=tips.query("size != 3"))
plt.show()
```

```python
sns.catplot(x="smoker", y="tip", 
            order=["No", "Yes"], 
            data=tips)
plt.show()
```

```python
sns.catplot(x="total_bill", y="day", 
            hue="time", kind="swarm", 
            data=tips)
plt.show()
```

# Boxplots

```python
sns.catplot(x="day", y="total_bill", 
            kind="box", data=tips)
plt.show()
```

```python
sns.catplot(x="day", y="total_bill", 
            hue="smoker", kind="box", 
            data=tips)
plt.show()
```

```python
tips["weekend"] = tips["day"].isin(["Sat", "Sun"])
sns.catplot(x="day", y="total_bill", 
            hue="weekend", kind="box", 
            dodge=False, data=tips)
plt.show()
```

```python
diamonds = sns.load_dataset("diamonds")
sns.catplot(x="color", y="price", 
            kind="boxen",
            data=diamonds.sort_values("color"))
plt.show()
```

# Violinplots

```python
sns.catplot(x="total_bill", y="day", 
            kind="violin", data=tips)
plt.show()
```

```python
sns.catplot(y="total_bill", x="day", 
            kind="violin", data=tips)
plt.show()
```

```python
sns.catplot(x="total_bill", y="day", 
            hue="time",
            kind="violin", data=tips)
plt.show()
```

```python
sns.catplot(x="total_bill", y="day", 
            hue="time", kind="violin", 
            bw=0.15, cut=0,
            data=tips)
plt.show()
```

```python
sns.catplot(x="day", y="total_bill", 
            hue="sex", kind="violin", 
            split=True, data=tips)
plt.show()
```

```python
sns.catplot(x="day", y="total_bill", 
            hue="sex", kind="violin", 
            inner="stick", split=True,
            palette="pastel", data=tips)
plt.show()
```

```python
g = sns.catplot(x="day", y="total_bill", 
            kind="violin", inner=None, 
            data=tips)
sns.swarmplot(x="day", y="total_bill", 
              color="k", size=3, 
              data=tips, ax=g.ax)
plt.show()
```

# Bar plots

```python
titanic = sns.load_dataset("titanic")
titanic
```

```python
sns.catplot(x="sex", y="survived", 
             kind="bar", 
             data=titanic)
plt.show()
```

```python
sns.catplot(y="sex", x="survived", 
             kind="bar", 
             data=titanic)
plt.show()
```

```python
sns.catplot(x="sex", y="survived", 
            hue="class", kind="bar", 
            data=titanic)
plt.show()
```

```python
sns.catplot(x="deck", kind="count", 
            palette="ch:0.25", 
            data=titanic)
plt.show()
```

```python
sns.catplot(y="deck", hue="class", 
            kind="count", palette="pastel", 
            edgecolor="0.6", data=titanic)
plt.show()
```

# Point plots

```python
sns.catplot(x="sex", y="survived", 
            kind="point", 
            data=titanic)
plt.show()
```

```python
sns.catplot(x="sex", y="survived", 
            hue="class", kind="point", 
            data=titanic)
plt.show()
```

```python
sns.catplot(x="class", y="survived", hue="sex",
            palette={"male": "g", "female": "m"},
            markers=["^", "o"], linestyles=["-", "--"],
            kind="point", data=titanic)
plt.show()
```

# Plotting â€œwide-formâ€ data

```python
iris = sns.load_dataset("iris")
iris
```

```python
sns.catplot(data=iris, 
            orient="h",
            kind="box")
plt.show()
```

```python
sns.violinplot(x=iris.species, 
               y=iris.sepal_length)
plt.show()
```

```python
f, ax = plt.subplots(figsize=(7, 3))
sns.countplot(y="deck", data=titanic, 
              color="c")
plt.show()
```

# Showing multiple relationships with facets

```python
sns.catplot(x="day", y="total_bill", 
            hue="smoker",
            col="time", aspect=.6,
            kind="swarm", data=tips)
plt.show()
```

```python
g = sns.catplot(x="fare", y="survived", 
                row="class", kind="box", 
                orient="h", height=1.5, aspect=4,
                data=titanic.query("fare > 0"))
g.set(xscale="log")
plt.show()
```
