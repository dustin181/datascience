# kernel Density estimation

```python
sns.distplot(x, hist=False, rug=True)
plt.show()
```

```python
x = np.random.normal(0, 1, size=30)
x
```

```python
sns.distplot(x, hist=False, rug=True)
plt.show()
```

```python
bandwidth = 1.06 * x.std() * x.size ** (-1/ 5.0)
support = np.linspace(-4, 4, 200)

kernels = []

for x_i in x:
    print(x)
    kernel = stats.norm(x_i, bandwidth).pdf(support)
    kernels.append(kernel)
    plt.plot(support, kernel, color="r")
    
sns.rugplot(x, color="0.2", linewidth = 3)
plt.show()
```

```python
from scipy.integrate import trapz
density = np.sum ( kernels, axis=0)
density= density / trapz(density, support)
plt.plot(support, density)
plt.show()
```

# Fitting parametric Distributions

```python
x = np.random.gamma(6, size=200)
sns.distplot(x, kde=False,
            fit=stats.gamma)
plt.show()
```

# Plotting bivariate distributions

```python
mean, cov = [0, 1], [(1, 0.5), (0.5, 1)]
data = np.random.multivariate_normal(mean, cov, 200)
df = pd.DataFrame(data, columns=["x", "y"])
df
```

# Scatter Plot

```python
sns.jointplot(x= "x", y="y", data=df)
plt.show()
```

# Hexbin Plot

```python
with sns.axes_style("white"):
    sns.jointplot(x="x", y="y",data= df, kind= "hex", color="g")
    plt.show()
```

# Kernel Density Estimation

```python
sns.jointplot(x="x", y="y", data=df, kind="kde")
plt.show()
```

```python
f, ax = plt.subplots(figsize=(6, 6))
sns.kdeplot(df.x, df.y, ax=ax)
sns.rugplot(df.x, color="g", ax=ax)
sns.rugplot(df.y, vertical=True, ax=ax)
plt.show()
```

```python
f, ax = plt.subplots(figsize=(6, 6))
cmap = sns.cubehelix_palette(as_cmap=True, dark=0,
                            light=1, reverse=True)
sns.kdeplot(df.x, df.y, cmap=cmap, n_levels=60, shade=True)
plt.show()
```

```python
g = sns.jointplot(x="x", y="y", data=df,
                 kind="kde", color="m")
g.plot_joint(plt.scatter, c="w", s=30,
            linewidth=1, marker="+")
g.ax_joint.collections[0].set_alpha(0)
g.set_axis_labels("$X$", "$Y$")
plt.show()
```

# Pairwise relationships

```python
iris = sns.load_dataset("iris")
sns.pairplot(iris)
plt.show()
```

```python
g= sns.PairGrid(iris)
g.map_diag(sns.kdeplot)
g.map_offdiag(sns.kdeplot, n_levels=6)
plt.show()
```
