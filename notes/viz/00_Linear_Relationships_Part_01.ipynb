```python
%matplotlib inline
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set(color_codes=True)
tips = sns.load_dataset("tips")
```

```python
tips
```

```python
sns.regplot(x="total_bill", y="tip", data=tips)
plt.show()
```

```python
sns.lmplot(x="total_bill", y="tip", data=tips)
plt.show()
```

```python
sns.lmplot(x="size", y="tip", data=tips)
plt.show()
```

```python
sns.lmplot(x="size", y="tip", 
           data=tips,
          x_jitter=0.05)
plt.show()
```

```python
sns.lmplot(x="size", y="tip", 
           data=tips,
          x_estimator=np.mean)
plt.show()
```

# fitting models

```python
anscombe = sns.load_dataset("anscombe")
```

```python
anscombe
```

```python
sns.lmplot(x="x", y="y",
          data=anscombe.query("dataset == 'I'"),
          ci=None, scatter_kws={"s": 80})
```

```python
sns.lmplot(x="x", y="y",
          data=anscombe.query("dataset == 'II'"),
          ci=None, scatter_kws={"s": 80})
```

```python
sns.lmplot(x="x", y="y",
          data=anscombe.query("dataset == 'II'"),
          ci=None, scatter_kws={"s": 80}, order=2)
```

```python
sns.lmplot(x="x", y="y",
          data=anscombe.query("dataset == 'III'"),
          ci=None, scatter_kws={"s": 80})
```

```python
sns.lmplot(x="x", y="y",
          data=anscombe.query("dataset == 'III'"),
          ci=None, scatter_kws={"s": 80}, robust=True)
```

```python
tips["big_tip"] = (tips.tip / tips.total_bill) > 0.15

tips
```

```python
sns.lmplot(x = "total_bill", y="big_tip", 
           data=tips, y_jitter=0.03)
plt.show()
```

```python
sns.lmplot(x = "total_bill", y="big_tip", 
           data=tips, logistic=True, 
           y_jitter=0.03)
plt.show()
```

```python
sns.lmplot(x = "total_bill", y="big_tip", 
           data=tips, lowess= True)
plt.show()
```

```python
sns.residplot(x="x", y="y",
          data=anscombe.query("dataset == 'I'"),
          scatter_kws={"s": 80})
plt.show()
```

```python
sns.residplot(x="x", y="y",
          data=anscombe.query("dataset == 'II'"),
          scatter_kws={"s": 80})
plt.show()
```

# Conditioning on other variables

```python
sns.lmplot(x="total_bill", y="tip",
          hue="smoker", data=tips)
plt.show()
```

```python
sns.lmplot(x="total_bill", y="tip",
          hue="smoker", data=tips,
          palette="Set1")
plt.show()
```

```python
sns.lmplot(x="total_bill", y="tip",
          hue="smoker", data=tips,
          palette="Set1",
          markers=["o", "x"])
plt.show()
```

```python
sns.lmplot(x="total_bill", y="tip",
          hue="smoker", data=tips,
          palette="Set1",
          markers=["o", "x"],
          col="time")
plt.show()
```

```python
sns.lmplot(x="total_bill", y="tip",
          hue="smoker", data=tips,
          palette="Set1",
          markers=["o", "x"],
          col="time", row="sex")
plt.show()
```

# control size and shape of the plot

```python
f, ax = plt.subplots(figsize=(5, 6))
sns.regplot(x="total_bill", y = "tip",
           data=tips, ax= ax)
plt.show()
```

```python
sns.lmplot(x="total_bill", y = "tip",
           data=tips, col="day",
          col_wrap=2, height=3)
plt.show()
```

```python
sns.lmplot(x="total_bill", y = "tip",
           data=tips, col="day",
          aspect=0.5)
plt.show()
```

# Plot regression in other context

```python
sns.jointplot(x="total_bill", y="tip",
             data=tips, kind="reg")
plt.show()
```

```python
sns.pairplot(tips, x_vars=["total_bill", "size"],
            y_vars=["tip"], height=5, aspect=0.8,
            kind="reg")
plt.show()
```

```python
sns.pairplot(tips, x_vars=["total_bill", "size"],
            y_vars=["tip"], height=5, aspect=0.8,
            kind="reg", hue="smoker")
plt.show()
```
