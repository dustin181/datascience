```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
```

# ScatterPlot

```python
tips = sns.load_dataset("tips")
```

```python
tips
```

```python
sns.relplot(x="total_bill", y ="tip",
           data=tips)
plt.show()
```

```python
sns.relplot(x="total_bill", y ="tip",
           data=tips, hue="smoker")
plt.show()
```

```python
sns.relplot(x="total_bill", y ="tip",
           data=tips, hue="smoker",
           style="smoker")
plt.show()
```

```python
sns.relplot(x="total_bill", y ="tip",
           data=tips, hue="smoker",
           style="time")
plt.show()
```

```python
sns.relplot(x="total_bill", y ="tip",
           data=tips, hue="size")
plt.show()
```

```python
sns.relplot(x="total_bill", y ="tip",
           data=tips, hue="size",
           palette="ch:r=-.5,l=.75")
plt.show()
```

```python
sns.relplot(x="total_bill", y ="tip",
           data=tips, size="size",
           palette="ch:r=-.5,l=.75")
plt.show()
```

```python
sns.relplot(x="total_bill", y ="tip",
           data=tips, size="size", 
            sizes=(10, 100), hue="size",
           palette="ch:r=-.5,l=.75")
plt.show()
```

# Line Plots

```python
df = pd.DataFrame(dict(time=np.arange(500),
                      value=np.random.randn(500).cumsum()))
g = sns.relplot(x="time", y="value", kind="line", data=df)
g.fig.autofmt_xdate()
plt.show()
```

```python
df = pd.DataFrame(np.random.randn(500, 2).cumsum(axis=0),
                 columns=["x", "y"])
sns.relplot(x="x", y="y", kind="line",
           sort=False, data=df)
plt.show()
```

# Aggregation and Representing Uncertainty

```python
fmri = sns.load_dataset("fmri")

fmri
```

```python
sns.relplot(x="timepoint", y="signal",
            kind="line", data=fmri)
plt.show()
```

```python
sns.relplot(x="timepoint", y="signal",
            kind="line", data=fmri, ci=None)
plt.show()
```

```python
sns.relplot(x="timepoint", y="signal",
            kind="line", data=fmri, ci="sd")
plt.show()
```

```python
sns.relplot(x="timepoint", y="signal",
            kind="line", data=fmri,
           estimator=None)
plt.show()
```

# Visualize subsets with semantic mappings

```python
sns.relplot(x="timepoint", y="signal",
            kind="line", data=fmri,
            hue="event")
plt.show()
```

```python
sns.relplot(x="timepoint", y="signal",
            kind="line", data=fmri,
            style="event", hue="region")
plt.show()
```

```python
sns.relplot(x="timepoint", y="signal",
            kind="line", data=fmri,
            style="event", hue="region",
           dashes=False, markers=True)
plt.show()
```

```python
sns.relplot(x="timepoint", y="signal",
            kind="line", data=fmri,
            style="event", hue="event")
plt.show()
```

```python
sns.relplot(x="timepoint", y="signal",
            kind="line", data=fmri.query("event == 'stim'"),
            hue="region", units="subject",
           estimator=None)
plt.show()
```

# Visualize Multiple Relationships

```python
tips = sns.load_dataset("tips")
sns.relplot(x="total_bill", y="tip",
           hue="smoker", col="time",
           data=tips)
plt.show()
```

```python
sns.relplot(x="timepoint", y="signal",
            kind="line", data=fmri, estimator = None,
            hue="subject", col="region", row= "event",
           height = 3)
plt.show()
```

```python
sns.relplot(x="timepoint", y="signal",
            hue="event", style="event",
            col = "subject",
           kind="line", data=fmri.query("region == 'frontal'"),
           height=3, aspect=0.75, linewidth=2.5)
plt.show()
```

# Time Series Plotting

```python
df = pd.DataFrame(dict(time=pd.date_range("2017-1-1", periods=500),
                      value=np.random.randn(500).cumsum()))
g = sns.relplot(x="time", y = "value", kind="line", data=df)
g.fig.autofmt_xdate()
plt.show()
```
